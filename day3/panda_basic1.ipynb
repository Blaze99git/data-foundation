```python
import pandas as pd

df = pd.read_csv("../day2/sample_data.csv")
df
```

```python
# How much did each customer spend on each category?
df.groupby(["customer", "category"])["amount"].sum()
```

```python
#What is the average order value per city? 
df.groupby(["city"])["amount"].mean()
```

```python
#Which category generates more revenue overall?
df.groupby(["category"])["amount"].sum().idxmax()
```
